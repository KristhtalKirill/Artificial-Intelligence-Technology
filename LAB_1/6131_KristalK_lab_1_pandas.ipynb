### 5 - Решейпинг данных с использованием .pivot (можно подать на вход результаты агрегатов, полученных ранее с использованием .groupby (сгруппировать по двум полям)

```python
df_resh = df.groupby(["Skewness", "SD_DMSNR_Curve"])["EK"].max().reset_index()
df_resh.pivot(index = "Skewness", columns = "SD_DMSNR_Curve", values= "EK")
df_resh
```
